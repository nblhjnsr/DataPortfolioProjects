```python
#import libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjusts the configuration of the plots we will create


#Read in the data. Gonna be using pandas to create dataframe to read csv

df = pd.read_csv(r'C:\Users\nabil\OneDrive\Desktop\Alex The Analyst Portfolio\4. Correlation in Python\movies.csv')
```

```python
# Let's look at the data

df.head()
```

```python
# See if there's any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data types for our columns

df.dtypes
```

```python
# change data type of columns

df['budget'] = df['budget'].apply("int64")

df['gross'] = df['gross'].apply("int64")
```

```python
df
```

```python
# Create correct year column

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)

df
```

```python
#Removing the missing values

df = df.dropna()

df
```

```python
# Check and see if there's any more missing values

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
pd.set_option('display.max_rows', None)
```

```python
df = df.sort_values(by=['gross'],inplace=False, ascending=False)
```

```python
#Drop any duplicates

df['company'].sort_values(ascending=False)
```

```python
# Budget high correlation 
```

```python
# scatter plot with budget vs gross
plt.scatter(x=df['budget'], y=df['gross']) 

plt.title('Budget vs Gross Earnings') 

plt.xlabel('Budget for Film') 

plt.ylabel('Gross Earnings') 

plt.show()
```

```python
correlation_matrix = df.corr(method='pearson')


plt.title('Correlation Matric for Numeric Features') 

plt.xlabel('Movie Features') 

plt.ylabel('Movie Features') 


sns.heatmap(correlation_matrix, annot=True)

plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
correlation_matrix = df_numerized.corr(method='pearson')


plt.title('Correlation Matric for Numeric Features') 

plt.xlabel('Movie Features') 

plt.ylabel('Movie Features') 


sns.heatmap(correlation_matrix, annot=True)

plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
