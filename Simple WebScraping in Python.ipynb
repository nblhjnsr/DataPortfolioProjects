```python
import requests
from bs4 import BeautifulSoup
import statistics
```

```python
BASE_URL = "https://books.toscrape.com/"

url = BASE_URL + "catalogue/category/books/philosophy_7/index.html"

#To get the content of the page into ebscraping the price and calculate the mean price of all the books in the page

response = requests.get(url)

soup = BeautifulSoup(response.text, "lxml")

#print(soup.find_all("p")) #find_all and "p" will give all the paragraphs

#print(soup.find_all("a")) #find_all and "a" will give all the anchors
```

```python
#Our goal is to get the price and the mean price. So we go to the website and right click + inspect to see the html of that page. 
#Then identify the code to the element we want, in this case the price

price_tags = soup.find_all("p", {"class": "price_color"})

#print(price_tags)



```

```python
#Get the price tags and filter them. [2:] is used to cut the unknown symbol at the beginning of the price tag
#Using the float function to convert the prices to float data type
#Get the mean price by 

prices = [float(price.text[2:]) for price in price_tags]

print(prices)

print(statistics.mean(prices))
```
